```python
import requests as requests
import urllib.request as url
```

```python
access_key = open(".access_key").read()
# print(access_key)
```

```python
target="forest"
```

```python
r = requests.get(f"https://api.unsplash.com/search/collections?query={target}&page=1&per_page=30&client_id={access_key}")
```

```python
data = r.json()
```

```python
for img_data in data['results'][:3]:
    file_name = str(img_data['id'])+".jpg"
    img_url = img_data['cover_photo']['urls']['raw']
    suffix = '&q=80&fm=jpg&crop=entropy&cs=tinysrgb&w=1920&fit=max'
    print(img_url)
    img_url = img_url + suffix
    url.urlretrieve(img_url, './result/'+file_name)
```
